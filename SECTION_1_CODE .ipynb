```python
import pandas as pd

df = pd.read_csv("hotel_bookings (1).csv", keep_default_na=False)

print(df.shape)
print(df.columns)
```

```python
property_city_count = (
    df.groupby('property_name')['property_city']
      .nunique()
      .reset_index(name='city_count')
)

duplicate_properties = property_city_count[
    property_city_count['city_count'] > 1
]

print(duplicate_properties)
```

```python
print("Number of property names appearing in multiple cities:")
print(len(duplicate_properties))
```

```python
for prop in duplicate_properties['property_name']:
    print("\n", "="*50)
    print("Property:", prop)

    display(
        df[df['property_name'] == prop]
        [['property_id','property_name','property_city']]
        .drop_duplicates()
        .sort_values('property_city')
    )
```

```python
print(df['coupon_code'].value_counts(dropna=False))
```

```python
cancelled_reviews = df[
    (df['booking_status'] == 'Cancelled') &
    (df['review_rating'] != '')
]

print("Cancelled bookings with reviews:", len(cancelled_reviews))
```
